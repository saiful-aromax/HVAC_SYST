```python
import pandas as pd
from datetime import datetime, timedelta

# Define the start date
start_date = datetime(2024, 3, 11)

# Create a list to hold the datetime objects
datetimes = [start_date + timedelta(minutes=30*x) for x in range(4*48)]

# Create a DataFrame
df = pd.DataFrame({
    "Date": [dt.date() for dt in datetimes],
    "Time": [dt.time() for dt in datetimes]
})


thermostat_set_temps = {
    "2024-03-11": 21,
    "2024-03-12": 20,
    "2024-03-13": 19,
    "2024-03-14": 20
}

# Update "Thermostat Set Temp" based on the date
df['Thermostat Set Temp'] = df['Date'].apply(lambda x: thermostat_set_temps[str(x)])

df['Thermostat Actual Temp'] = 18.0

for i in range(len(df)):
    if i > 0 and df.loc[i, 'Date'] == df.loc[i-1, 'Date']:
        if df.loc[i-1, 'Thermostat Actual Temp'] < df.loc[i-1, 'Thermostat Set Temp']:
            df.loc[i, 'Thermostat Actual Temp'] = df.loc[i-1, 'Thermostat Actual Temp'] + 0.5
        else:
            df.loc[i, 'Thermostat Actual Temp'] = df.loc[i-1, 'Thermostat Actual Temp'] - 0.5
    else:
        df.loc[i, 'Thermostat Actual Temp'] = 18.0  # Reset for a new day
df['Humidity'] = 60 - 2 * (df['Thermostat Actual Temp'] - 18)
```

```python
# Define TRV Set Temp for each day
trv_set_temps = {
    "2024-03-11": 22,
    "2024-03-12": 23,
    "2024-03-13": 18,
    "2024-03-14": 19
}

# Add "TRV Set Temp" based on the date
df['TRV1 Set Temp'] = df['Date'].apply(lambda x: trv_set_temps[str(x)])

# Initialize "TRV Actual Temp" with the same starting conditions as before
df['TRV1 Actual Temp'] = 18.0

# Update "TRV Actual Temp" values using the same logic as before
for i in range(len(df)):
    if i > 0 and df.loc[i, 'Date'] == df.loc[i-1, 'Date']:
        if df.loc[i-1, 'TRV1 Actual Temp'] < df.loc[i-1, 'TRV1 Set Temp']:
            df.loc[i, 'TRV1 Actual Temp'] = df.loc[i-1, 'TRV1 Actual Temp'] + 0.5
        else:
            df.loc[i, 'TRV1 Actual Temp'] = df.loc[i-1, 'TRV1 Actual Temp'] - 0.5
    else:
        df.loc[i, 'TRV1 Actual Temp'] = 18.0  # Reset for a new day

# Add "TRV Humidity" based on "TRV Actual Temp" using the same logic as before
df['TRV1 Humidity'] = 60 - 2 * (df['TRV1 Actual Temp'] - 18)

df.head(10)  # Show the first 10 rows to illustrate the addition
```
